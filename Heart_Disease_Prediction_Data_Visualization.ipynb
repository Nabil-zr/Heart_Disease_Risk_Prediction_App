**1. DATA LOAD**

```python
import pandas as pd

# Replace 'filename.csv' with the actual name of your uploaded file
data = pd.read_csv('my_clean_data.csv')
data.head()
```

**2. DATA VISUALIZATION**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set visual style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (6, 3)

# 1. Heart Disease Class Distribution
sns.countplot(data, x='HeartDisease', palette='Set2')
plt.title("Heart Disease Distribution")
plt.xlabel("Heart Disease")
plt.ylabel("Count")
plt.show()
```

```python
# 2. BMI Distribution by Heart Disease

for label, color in zip(['Yes', 'No'], ['red', 'green']):
    subset = data[data['HeartDisease'] == label]
    sns.kdeplot(subset['BMI'], fill=True, label=f'Heart Disease: {label}', color=color)
plt.title("BMI Distribution by Heart Disease")
plt.xlabel("BMI")
plt.legend()
plt.show()
```

```python
# 3. Heart Disease by Age Category
sns.countplot(data, y='AgeCategory', hue='HeartDisease', palette='Set2')
plt.title("Heart Disease Count by Age Category")
plt.xlabel("Count")
plt.ylabel("Age Category")
plt.legend(title='Heart Disease')
plt.show()
```

```python
# 4. Heart Disease by Sex and Age

sns.catplot(data, x='AgeCategory', hue='HeartDisease', col='Sex',
            kind='count', col_order=['Male', 'Female'], palette='Set1', height=5, aspect=1.5)
plt.xticks(rotation=45)
plt.subplots_adjust(top=0.85)
plt.suptitle("Heart Disease by Sex and Age Category")
plt.show()
```

```python
# 5. Correlation Heatmap (Numerical features)
num_cols = ['BMI', 'PhysicalHealth', 'MentalHealth', 'SleepTime']
corr = data[num_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Numerical Features")
plt.show()
```

```python
# 6. Categorical Features vs Heart Disease
cat_features = ['Smoking', 'AlcoholDrinking', 'Stroke', 'DiffWalking', 'Diabetic',
                'PhysicalActivity', 'Asthma', 'KidneyDisease', 'SkinCancer']

for col in cat_features:
    sns.countplot(data, x=col, hue='HeartDisease', palette='Set2')
    plt.title(f"{col} vs Heart Disease")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# 7. Sleep Time vs General Health

sns.boxplot(data, x='GenHealth', y='SleepTime', hue='HeartDisease', palette='Set3')
plt.title("Sleep Time by General Health and Heart Disease")
plt.xlabel("General Health")
plt.ylabel("Sleep Time")
plt.legend(title="Heart Disease")
plt.show()
```
